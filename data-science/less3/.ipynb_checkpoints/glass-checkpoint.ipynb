```python
!pip install seaborn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('./glass.csv')
df.head()

features = [
'ID', 
'RI', 
'NA2O',
'MGO', 
'AL2O3', 
'SIO2', 
'K2O', 
'CAO', 
'BAO', 
'FE2O3',
'TYPE']


df = pd.DataFrame(df.values, columns=features)
# df.head()

df = df.drop('ID', axis= 1)

df.describe()
```

```python
data = df.drop('TYPE', axis= 1).values
target = df['TYPE'].values
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier(n_neighbors=1)
from sklearn.model_selection import KFold

errors = 0
kf = KFold(n_splits=5, shuffle=True)
for tr, ts in kf.split(data):
    kn.fit(data[tr], target[tr])

    errors += sum(target[ts] != kn.predict(data[ts]))
    
quality = (1 - errors/len(data))*100

quality
```

```python
his = np.histogram(target, bins=7)
plt.bar(his[1][1:], his[0])
```

```python
sns.distplot(target)
```

```python
from sklearn.model_selection import LeaveOneOut

loo = LeaveOneOut()
def get_precision(data, model):
    errors = 0
    for tr, ts in loo.split(data):
        model.fit(data[tr], target[tr])
        errors += sum(target[ts] != model.predict(data[ts]))


    quality = (1 - errors/len(data))*100
    return quality

get_precision(data, kn)
```

```python
for n in range(1, 10):
    kn = KNeighborsClassifier(n_neighbors=n)
    print(get_precision(data, kn))
```
