```python
from sklearn.svm import SVC, LinearSVC
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# from sklearn.svm 
import matplotlib.pyplot as plt

import numpy as np

iris = load_iris()
data = iris.data
target = iris.target


def graph(data, target, kn):
    x_min, x_max = data[:,0].min() - 0.5, data[:,0].max() + 0.5
    y_min, y_max = data[:,1].min() - 0.5, data[:,1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

    kn.fit(data, target)

    z = kn.predict(np.c_[xx.ravel(), yy.ravel()])

    z = z.reshape(xx.shape)
        
    plt.pcolormesh(xx, yy, z)
    plt.scatter(data[:,0][target == 0], data[:,1][target == 0], marker='+')
    plt.scatter(data[:,0][target == 1], data[:,1][target == 1], marker='+')
    plt.scatter(data[:,0][target == 2], data[:,1][target == 2], marker='+')
    plt.show()
    
dtr, dts, ttr, tts = train_test_split(data, target)

sc = SVC(kernel='linear')
sc.fit(dtr, ttr)


accuracy_score(tts, sc.predict(dts))
# graph(data[:, :2], target, sc)
```

```python
sc = LinearSVC()
sc.fit(dtr, ttr)




accuracy_score(tts, sc.predict(dts))
# graph(data[:, :2], target, sc)
```

```python
for i in (1, 10, 100):
#     plt.figure()
    plt.title("index%d"%i)
    graph(data[:, :2], target, SVC(kernel='linear', C=i))
```

```python
for i in (1, 10, 100):
#     plt.figure()
    plt.title("index%d"%accuracy_score(tts, sc.predict(dts)))
    graph(data[:, :2], target, SVC(kernel='rbf', C=i))
```

```python
for i in (1, 10, 100):
#     plt.figure()
    plt.title("index%d"%accuracy_score(tts, sc.predict(dts)))
    graph(data[:, :2], target, SVC(kernel='poly', C=i))
```

```python
for i in (1, 10, 100):
#     plt.figure()
    plt.title("index%d"%accuracy_score(tts, sc.predict(dts)))
    graph(data[:, :2], target, SVC(kernel='sigmoid', C=i))
```
