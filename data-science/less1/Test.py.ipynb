```python
df_kiev = df[(df['City'] == 'Kiev')][['dt', 'AverageTemperature', 'AverageTemperatureUncertainty']]
```

```python
df_kiev.plot()
```

```python
df_kiev_full = df_kiev.fillna(df_kiev['AverageTemperature'].median())
df_kiev_full['year'] = df_kiev_full['dt'].apply(lambda x: pd.to_datetime(x).year)
```

```python
df_kiev_full.groupby('year')['AverageTemperature'].mean().plot()
```

```python
df_kiev_full['year5']= df_kiev_full['year'].apply(lambda x: x // 5 * 5)
```

```python
df_kiev_full.groupby('year5')['AverageTemperature'].mean().plot()
```

```python
years = df_kiev_full['year'].values
temps = df_kiev_full['AverageTemperature'].values
z = np.polyfit(years, temps, 5)
temps
```

```python
f1 = np.poly1d(z)
```

```python
plt.plot(years, f1(years))
```

```python
np.mean(np.abs(temps - f1(years)))
```
