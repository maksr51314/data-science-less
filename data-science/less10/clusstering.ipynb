```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

iris = load_iris()

data = iris.data
target = iris.target

kn = KMeans(n_clusters=3)

kn.fit(data)

kn.cluster_centers_
kn.labels_

# plt.scatter(data[:,2][kn.labels_ == 0], data[:,3][kn.labels_ == 0], marker='o')
# plt.scatter(data[:,2][kn.labels_ == 1], data[:,3][kn.labels_ == 1], marker='o')
plt.scatter(data[:,2][kn.labels_ == 2], data[:,3][kn.labels_ == 2], marker='o')

# plt.scatter(data[:,2][target == 0], data[:,3][target == 0], marker='+')
plt.scatter(data[:,2][target == 1], data[:,3][target == 1], marker='+')
# plt.scatter(data[:,2][target == 2], data[:,3][target == 2], marker='+')
```

```python
dist = []

for i in range(1, 10):
    km = KMeans(n_clusters=i)
    km.fit(data)
    dist.append(km.inertia_)
dist

plt.plot(range(1, 10), dist)
```
