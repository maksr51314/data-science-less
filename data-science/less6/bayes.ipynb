```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import GaussianNB
import stop_words

df = pd.read_csv("normal.csv")
data = CountVectorizer().fit_transform(df["comment"])

stw = stop_words.get_stop_words('ru')
stw
# stw.remove('не')
# stw
```

```python
df['comment'] = df['comment'].apply(lambda x: " ".join(word for word in x.split() if word not in stw))

data = CountVectorizer(min_df=10).fit_transform(df.comment)
cv = CountVectorizer(min_df=10).fit(df.comment)
cv.transform(df.comment)
# data
cv
```

```python
cv.transform(["очень плохо"])
```

```python
gn = GaussianNB()

gn.fit(data.toarray(), df.target.values)
```

```python
gn.predict(cv.transform(['плохой']).toarray())
```

```python
gn.predict(cv.transform(['хороший']).toarray())
```

```python
from sklearn.model_selection import train_test_split
dtr, dts, ttr, tts = train_test_split(data.toarray(), df.target)

gn.fit(dtr, ttr)
from sklearn.metrics import accuracy_score

accuracy_score(tts, gn.predict(dts))
```

```python
df1 = pd.read_csv("normal.csv")
cv1 = CountVectorizer(min_df=10)
cv1.fit(df1.comment)
data = cv1.transform(df1.comment)
dtr, dts, ttr, tts = train_test_split(data.toarray(), df.target)
gn.fit(dtr, ttr)
accuracy_score(tts, gn.predict(dts))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(min_df=10)
data = tf.fit_transform(df.comment)

dtr, dts, ttr, tts = train_test_split(data.toarray(), df.target)
gn.fit(dtr, ttr)
accuracy_score(tts, gn.predict(dts))
```

```python
tf = TfidfVectorizer(min_df=50)
data = tf.fit_transform(df.comment)

dtr, dts, ttr, tts = train_test_split(data.toarray(), df.target)
gn.fit(dtr, ttr)
accuracy_score(tts, gn.predict(dts))
```
